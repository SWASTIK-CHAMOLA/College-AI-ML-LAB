```python
# Lab 5: Exploratory Data Analysis on Netflix Dataset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
file_path = "/content/drive/MyDrive/Colab Notebooks/Netflix Dataset.csv"
df = pd.read_csv(file_path)

print("Dataset Shape:", df.shape)
print("\nColumn Names:", df.columns.tolist())
print("\nDataset Info:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe(include="all").transpose())

# -------------------------------
# 1. Distribution of Categories (Movie vs TV Show)
# -------------------------------
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Category", palette="Set2")
plt.title("Distribution of Movies vs TV Shows")
plt.show()

# -------------------------------
# 2. Top 10 Countries with Most Content
# -------------------------------
plt.figure(figsize=(10,6))
df['Country'].value_counts().head(10).plot(kind='bar', color="skyblue")
plt.title("Top 10 Countries with Most Content")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()

# -------------------------------
# 3. Ratings Distribution
# -------------------------------
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Rating", order=df['Rating'].value_counts().index, palette="viridis")
plt.title("Distribution of Ratings")
plt.xticks(rotation=45)
plt.show()

# -------------------------------
# 4. Release Year Trend
# -------------------------------
# Extract year from Release_Date
df['Year'] = pd.to_datetime(df['Release_Date'], errors='coerce').dt.year

plt.figure(figsize=(12,6))
df['Year'].value_counts().sort_index().plot(kind="line", marker="o")
plt.title("Number of Releases Over the Years")
plt.xlabel("Year")
plt.ylabel("Count of Shows/Movies")
plt.grid(True)
plt.show()

# -------------------------------
# 5. Duration Distribution (Movies Only)
# -------------------------------
movie_df = df[df['Category']=="Movie"]

plt.figure(figsize=(8,5))
sns.histplot(movie_df['Duration'].str.replace(" min","").astype(float), bins=30, kde=True, color="red")
plt.title("Distribution of Movie Durations")
plt.xlabel("Duration (Minutes)")
plt.ylabel("Frequency")
plt.show()

# -------------------------------
# 6. Most Common Genres
# -------------------------------
plt.figure(figsize=(12,6))
df['Type'].value_counts().head(10).plot(kind='bar', color="orange")
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
```
